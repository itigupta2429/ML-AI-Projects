## Read data from csv file and perform preprocessing

```python
import pandas as pd
df = pd.read_csv("Social_Network_Ads.csv")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 10)
```

```python
df
```

```python
# Check for null/missing values:
df = df.dropna()

# Check for and fill missing values:
df = df.fillna(df.mean(numeric_only=True))

# Drop duplicate columns
df = df.loc[:, ~df.T.duplicated()]

# Drop constant value columns
df = df.loc[:, df.nunique() > 1]

# No changes to this data, as it's a pretty clean dataset
df.shape
```

#### Any encoding required?

No, the Prediction column (Purchased) is already encoded, if it wasn't, follow this step: 

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["Purchased"] = label_encoder.fit_transform(df["Purchased"])
df.head(8)
```

and then encode any other categorical columns - We have none here.

```python
# Let's take a look at our final dataset:
import matplotlib.pyplot as plt
import seaborn as sn

sn.pairplot(data = df, hue = "Purchased" )
```

## Create training and testing sets

```python
# Create X and y data
x_columns = 2
X = df.iloc[:, 0:x_columns].values
y = df.iloc[:, x_columns].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
print(len(X_train),len(y_train),len(X_test),len(y_test))
```

## Create Scaled Training and Testing Data

```python
# Retain unscaled test data for visualization later:
X_test_unscaled = X_test.copy()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(len(X_train), len(X_test))
print(X_train[0], X_test[0])
```

```python
## Set-up 'Result' visualization:

def plotResults(predictions):
    plt.figure(figsize=(10, 3))
    sn.scatterplot(data=df, x="Age", y="EstimatedSalary", hue="Purchased", palette="coolwarm", s=100, edgecolor="black", label="Actual")
    
    # Overlay predictions with different markers
    sn.scatterplot(x= X_test_unscaled[:,0], y=X_test_unscaled[:,1], hue=predictions, palette="coolwarm",  s=50, marker="X", edgecolor="black", label="Predicted")
    
    # Labels and title
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("Classification Scatter Plot: Actual vs. Predictions")
    plt.legend()
    plt.show()
```

## Logistic Regression Estimator

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression(solver = 'lbfgs', 
                           multi_class='auto', # use â€˜multinomialâ€™ if more than 2 classes, default=â€™autoâ€™, which will interpert binary or multi
                           max_iter=1000).fit(X_train,y_train)
predictions_LR = model_LR.predict(X_test)

print("Model Score = ",model_LR.score(X_test,y_test))
plotResults(predictions_LR)
```

## Support Vector Machines (SVM) Classifier

```python
from sklearn.svm import SVC

# Create an instance of 'svm_model'
model_SVM = SVC(max_iter = 1000,
               kernel = "rbf") # max_iter is not required, but can prevent crashing when models don't converge
model_SVM.fit(X_train, y_train) 

predictions_SVM = model_SVM.predict(X_test)
print("Model Score = ",model_SVM.score(X_test,y_test))
plotResults(predictions_SVM)
```

### Optimize a SVC Model using GridSearch

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01,0.001,0.0001]}
```

```python
# We need to import GridSearchCV from model_selection. ** Must be capitalized in GridSearchCV! **
from sklearn.model_selection import GridSearchCV
```

```python
# refit = True by default, this re-trains the best model and saves it in 'grid' for further use. 
SVC_grid = GridSearchCV(SVC(), param_grid, verbose=2, refit = True) 

#grid is an instance of the gridsearch. 
```

```python
# Run a model and save the results for each combination of variables in the grid
SVC_grid.fit(X_train,y_train)
```

```python
print('The best parameters are %s with a score of %0.2f' 
      % (SVC_grid.best_params_, SVC_grid.best_score_))
```

```python
SVC_grid.best_params_
```

```python
#Now we have everything in grid, including the best one, so we can just use that:
grid_SVC_predictions = SVC_grid.predict(X_test)

print("Model Score = ",SVC_grid.score(X_test,y_test))
plotResults(grid_SVC_predictions)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

model_tree = DecisionTreeClassifier(criterion = "entropy", 
                                    max_leaf_nodes = 1000)
model_tree.fit(X_train, y_train)

predictions_tree = model_tree.predict(X_test)
print("Model score = ",model_tree.score(X_test,y_test))

plotResults(predictions_tree)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RF = RandomForestClassifier(n_estimators = 10, # Default is 100
                                  bootstrap = True) # If false, then all features will be used in each tree ("bagging")
                                                    # class_weight allows you to add weighting ("boosting")   
model_RF.fit(X_train, y_train)

predictions_RF = model_RF.predict(X_test)
print("Model Score =",model_RF.score(X_test,y_test))
plotResults(predictions_RF)
```

## Naive Bayes Classifier

From ScikitLearn - These classifiers are based on applying Bayesâ€™ theorem with the â€œnaiveâ€ assumption of conditional independence between every pair of features given the value of the class variable.
i.e. It assumes the presence of a particular feature in a class is unrelated to the presence of any other feature.

In spite of their apparently over-simplified assumptions, naive Bayes classifiers have worked quite well in many real-world situations, famously document classification and spam filtering. They require a small amount of training data to estimate the necessary parameters.

They do well at guessing the label (a good classifier), but it is known to be a bad estimator - so don't rely on the specific probability. 

```python
from sklearn.naive_bayes import GaussianNB
model_NB = GaussianNB().fit(X_train, y_train)

predictions_NB = model_NB.predict(X_test)
print("Model Score = ", model_NB.score(X_test,y_test))
plotResults(predictions_NB)
```

## Now choose the best model and predict!

```python
#Random forest and the Optimized SVC had the best score, so lets choose Random Forest. 

# Let's pretend a new custom saw our add, will they make a purchase?

# Create an array for our new customer, aged 27 and making approx. $56,000:
import numpy as np
new_customer = np.array([[27, 56000]])

# Use the scaler we determined above on teh training data to scale our new customer data:
new_customer = scaler.transform(new_customer)

# Use the newly scaled data in our chosen model to make a prediction!
prediction = model_RF.predict(new_customer)
print(prediction)
```
