```python
### Grouping data in dataframe
mushroom_data.groupby(by="class").count()
```

```python
mushroom_data.groupby(["class","odor"]).count()
```

```python
## Filter data on a column
mushroom_edible=mushroom_data[(mushroom_data["class"] == "e") & (mushroom_data["odor"] == "a")]
```

```python
##create subset of a dataframe
mushroom_subset=mushroom_data[["class", "odor"]]
```

```python
##Concat 2 df
pd.concat([mushroom_edible,mushroom_subset])
```
